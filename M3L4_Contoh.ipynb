```python
''' mengikuti kursus persiapan meningkatkan hasil ujian bagi pelamar dari keluarga yang kedua orang tuanya tidak memiliki pendidikan tinggi '''

# import library
import pandas as pd
# bikin dataframe
df = pd.read_csv('StudentsPerformance .csv')


'''1. apakah tingkat pendidikan ortu mempengaruhi nilai'''
# bikin kolom baru untuk rata2 nilai
df['average'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

level = df['parental level of education'].value_counts()
# print(level)

# rata2 nilai untuk tiap level pendidikan ortu
data = df.groupby(by='parental level of education')['average'].mean()
print(data)

# hipotesis terbukti


'''2. apakah kalau ikut kursus nilai lebih tinggi'''
data2 = df.groupby(by='test preparation course')['average'].mean()
print(data2)

# hipotesis terbukti


'''3. apakah ikut kursus bisa menaikkan nilai yang ortunya pendidikan tidak tinggi'''

# pake groupby()
data3 = df.groupby(by = ['parental level of education', 'test preparation course'])['average'].mean()
print(data3)

data_pivot = df.pivot_table(
    index='parental level of education',
    columns='test preparation course',
    values='average',
    aggfunc=['mean']
    )

print(data_pivot)

# per pelajaran
data_pivot = df.pivot_table(
    index='parental level of education',
    columns='test preparation course',
    values=['math score', 'reading score', 'writing score'],
    aggfunc=['mean']
    )

print(data_pivot)

# hipotesis terbukti

# print(df.info())
# print(df.head())
```
