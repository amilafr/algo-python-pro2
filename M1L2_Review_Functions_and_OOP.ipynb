```python
class Book():
  def __init__(self, title, genre): # constructor
    # property
    self.title = title
    self.genre = genre

  # method
  def print_book(self):
    print('This is', self.title)

# bikin objek
horror_book = Book('The Conjuring', 'horror')

horror_book.print_book()
```

```python
print(horror_book.genre)
```

```python
# class --> general description
class Car():
  def __init__(self, brand, color): # constructor
    # property
    self.brand = brand
    self.color = color

  # method
  def print_data(self):
    print('Brand:', self.brand)
    print('Color:', self.color)

# object
red_car = Car(color='red', brand='toyota')

# calling method
red_car.print_data()

# calling property
brand = red_car.brand
print(brand)
```

```python
class PriceList():
    def __init__ (self, name):
        self.name = name
        self.pricelist = dict()
    def add_price(self, product, price):
        self.pricelist[product] = price

my_offer = PriceList('Services for Instagram')
my_offer.add_price('stories', 100)
```

```python
def data(d):
  for key in d:
    print(key, '-', d[key])

d = {'name': 'Rakha', 'age': 16}

data(d)
```

```python
def data(**dataa):
  for key in dataa:
    print(key, '-', dataa[key])

data(name='rakha', age=16, city='jakarta')
```

```python
def add_price(**kwargs):
    for key in kwargs:
        print('Service –',key,', price –',kwargs[key])

add_price(stories = 100, management = 1000)
```

```python
class Offer():
    def add_price(self ,**kwargs):
        for key in kwargs:
            print('Service –',key,', price –',kwargs[key])

my_offer = Offer()
my_offer.add_price(stories = 100, management = 1000)
```

```python
class Price_list():
    def __init__ (self,name):
        self.name = name
        self.pricelist = dict()

    # argument --> dictionary
    def add_prices(self, **kwargs):
        for key in kwargs:
            self.pricelist[key] = kwargs[key]

my_offer = Price_list('Instagram')
my_offer.add_prices(stories = 100, management = 1000)
```

# Solutions

```python
#class Price_list():
class Price_list():
   def __init__ (self,name):
       self.name = name
       self.pricelist = dict()
   def add_price(self, **kwargs):
       for key in kwargs:
           self.pricelist[key] = kwargs[key]
   def order(self, **kwargs):
       sum = 0
       for key in kwargs:
           if key in self.pricelist:
               sum += self.pricelist[key] * kwargs[key]
       return sum

my_offer = Price_list('Instagram')
my_offer.add_price(management = 1000, content_plan = 850, style = 500, stories = 100, post = 300)

while input('Would you like to make an order? (1 - yes, 0 - no)') != '0':
    operation = input('Would you like to order management of your accounts (1) or posts (2)?')
    if operation == '1':
        management = int(input('How many new accounts would you like to add?'))
        content_plan = int(input('For how many of those will we create a content plan?'))
        style = int(input('For how many of those will we develop a style?'))
        price = my_offer.order(management = management, content_plan = content_plan, style = style)
    else:
        stories = int(input('How many stories would you like to order?'))
        post = int(input('How many posts would you like to order?'))
        price = my_offer.order(stories = stories, post = post)
    print(f'Service cost: {price} coins.')

print('Thank you for working with us! Have a nice day!')

# my_offer = Price_list('Instagram')
# my_offer.add_price(management = 1000, content_plan = 850, style = 500, stories = 100, post = 300)
```
