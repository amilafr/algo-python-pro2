## Scatter Plot

image.png

**Constructing Scatter Plot**

- Select the two attributes we need to visualize the relationship between.
- Mark the points on the coordinate plane whose coordinates represent the values of the selected attributes.
- Evaluate the position of the points on the graph: are they scattered evenly along the coordinate plane, or do they cluster around an imaginary line?
- If the points are clustered around an imaginary line, there is a connection between the quantities; if they are not, there is no connection.

```python
#write your code here
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('StudentsPerformance.csv')

# df.info()
df.head()
```

```python
df['mean score'] = (df['math score'] + df['reading score'] + df['writing score']) / 3

df.head()
```

## linear graph

```python
df['mean score'].plot(kind='line')
```

## histogram

```python
df['math score'].plot(kind='hist') # persebaran data math score
```

```python
df['reading score'].plot(kind='hist')
```

```python
df['writing score'].plot(kind='hist')
```

```python
df['mean score'].plot(kind='hist')
```

## boxplot

```python
df[df['test preparation course'] == 'completed']['mean score'].plot(kind = 'box') # median, dll nya mean score yang test = completed
```

```python
df[df['test preparation course'] == 'none']['mean score'].plot(kind = 'box')
```

## scatter plot

```python
df.plot(x = 'math score', y = 'reading score', kind = 'scatter')
# hubungannya math score sama reading score --> yg math score nya tinggi biasanya readingnya juga tinggi
```

```python
df.plot(x = 'math score', y = 'writing score', kind = 'scatter')
```

## pie chart

```python
df['parental level of education'].value_counts().plot(kind = 'pie')
```

```python
df['test preparation course'].value_counts().plot(kind = 'pie')
```

## bar chart

```python
# rata2 score per test preparation dan parental
d = df.pivot_table(
    index = 'parental level of education',
    columns = 'test preparation course',
    values = 'mean score',
    aggfunc = 'mean'
    )

d
```

```python
d.plot(kind = 'barh', grid = True)
```
