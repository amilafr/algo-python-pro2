```
<!DOCTYPE html>
<html>

<head>
  <title>haii</title>
</head>

<body>
  <h1 style="color: red;">
    Write and run HTML, CSS and JavaScript code online.
  </h1>
  
  <p>
    Our HTML editor updates the webview automatically in real-time as you write code.
  </p>
</body>

</html>
```

### In head tag

image.png



```
<!DOCTYPE html>
<html>

<head>
  <title>haii</title>
  <style>
    h1{color: red;}
    p{color: blue;}
  </style>
</head>

<body>
  <h1>
    Write and run HTML, CSS and JavaScript code online.
  </h1>
  
  <p>
    Our HTML editor updates the webview automatically in real-time as you write code.
  </p>
</body>

</html>
```
