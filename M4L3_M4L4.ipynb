# Example!

```python
#create your individual project here!

# import library
import pandas as pd

from sklearn.model_selection import train_test_split # bagi data train-test
from sklearn.preprocessing import StandardScaler # standarin metrics
from sklearn.neighbors import KNeighborsClassifier # knn
from sklearn.metrics import confusion_matrix, accuracy_score # evaluasi


# dataframe
df = pd.read_csv('train.csv')

'''Clean data'''

# hapus kolom yang tidak perlu
df = df.drop([
    'id', 'bdate', 'langs', 'city', 'last_seen',
    'occupation_name', 'life_main', 'people_main',
    'career_start', 'career_end'
    ], axis=1)

# isi data kosong
# education_form --> yang kosong diisi Full Time
df['education_form'].fillna('Full-time', inplace=True)

# occupation_type --> yang kosong diisi university
df['occupation_type'].fillna('university', inplace=True)

# change data type
# education form
df[list(pd.get_dummies(df['education_form']).columns)] = pd.get_dummies(df['education_form'])
df = df.drop(['education_form'], axis=1)

# education status
df[list(pd.get_dummies(df['education_status']).columns)] = pd.get_dummies(df['education_status'])
df = df.drop(['education_status'], axis=1)

# occupation
def change_occupation(occ):
    if occ == 'university':
        return 0
    else:
        return 1

df['occupation_type'] = df['occupation_type'].apply(change_occupation)

# print(df.head())
print(df.info())

'''Build KNN'''

# pisahin x dan y
# x --> parameter
# y --> target
x = df.drop('result', axis=1)
y = df['result']

# split data --> train 80% , test 20%
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2)

# standardize metrics
sc = StandardScaler()
train_x = sc.fit_transform(train_x)
test_x = sc.transform(test_x)

# build knn model
model = KNeighborsClassifier(n_neighbors=3) # jumlah neighbors = 3

# training process
model.fit(train_x, train_y)

# testing process

# prediksi
y_pred = model.predict(test_x)

# evaluasi

acc = accuracy_score(test_y, y_pred)
print(acc)

# confusion matrix
cm = confusion_matrix(test_y, y_pred)
print(cm)
```
