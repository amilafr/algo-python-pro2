## Data Filtering

```python
'''
Determine the average number of downloads (Installs) of applications
whose rating ( Rating) exceeds the value of 4.9.

--> Rata2 jumlah install app yg rating > 4.9
'''

# data about apps with a rating greater than 4.9.
df[df['Rating'] > 4.9]
```

```python
# Install-nya aja
df[df['Rating'] > 4.9]['Installs']
```

```python
# rata2
df[df['Rating'] > 4.9]['Installs'].mean()
# rata2 install yang rating > 4.9
```

```python
print('Rata2 jumlah install app yg rating > 4.9')
print(df[df['Rating'] > 4.9]['Installs'].mean())
```

```python
'''
Determine the average number of downloads (Installs)
of free apps (Type == 'Free') with a rating (Rating) exceeding 4.9.
'''

# Rata2 install free app yang rating > 4.9
df[(df['Rating'] > 4.9) & (df['Type'] == 'Free')]['Installs'].mean()

# rata2 install yang rating > 4.9 & free apps
```

## No. 2

```python
import pandas as pd

df = pd.read_csv('GoogleApps.csv')

print(df.info())

# What is the price of the cheapest paid app (type == 'Paid')?
print('Paid app termurah')
print(df[df['Type'] == 'Paid']['Price'].min())

# What is the median number of installs
# for the applications from the "ART_AND_DESIGN" category?
print('Median jumlah install art and design category')
print(df[df['Category'] == 'ART_AND_DESIGN']['Installs'].median())

# How much more is the maximum number of Reviews for free apps (type == 'Free')
# than the maximum number of Reviews for paid apps (Type == 'Paid')?
print('Selisih max review dari free dan paid app')
free = df[df['Type'] == 'Free']['Reviews'].max()
paid = df[df['Type'] == 'Paid']['Reviews'].max()
print(free - paid)

# What is the minimum size of an application for teenagers (Content Rating == 'Teen')?
print('Minimum size teen app')
print(df[df['Content Rating'] == 'Teen']['Size'].min())

# *What is the category of an app having the most number of reviews?
print('Category yang punya review paling banyak')
print(df[df['Reviews'] == df['Reviews'].max()]['Category'])

# *What is the mean rating of applications which price exceeds $20 and
# the number of installs is over 10,000?
print('Mean rating yang price>20 dan install>10000')
print(df[(df['Price'] > 20) & (df['Installs'] > 10000)]['Rating'].mean())
```
