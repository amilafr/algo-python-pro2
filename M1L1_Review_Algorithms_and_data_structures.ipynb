### while loop

```python
# while --> perulangan tidak terbatas (selama kondisi benar)

ans = input('Insert code (off - stop): ').lower()

while ans != 'off': # selama ans != 'off' --> ngulang terus
  print(ans)
  ans = input('Insert code (off - stop): ').lower()
```

### example

```python
# split string jadi list
income = input('Enter monthly income data').split()

# semua elemen di list income jadi integer
income = list(map(int, income))

total = 0

minimum = income[0]
maximum = income[0]

for item in income:
    total += item
    if item > maximum:
        maximum = item
    if item < minimum:
        minimum = item

print('Total earnings:', total)
print('Maximum earnings:', maximum)
print('Minimum earnings:', minimum)
```

### example

```python
data = dict()
amount = int(input('Enter the number of startups'))

for i in range(amount):
    name = input('Enter startup name')
    data[name] = dict()
    data[name]['founder'] = input('Enter founder name')
    data[name]['investments'] = int(input('Investments attracted:'))

max_investments = 0

for startup in data:
    if data[startup]['investments'] > max_investments:
        max_investments = data[startup]['investments']
        winner = startup

print('Most investments attracted by', winner)
print('Founder:', data[winner]['founder'])
print('Total investments:', data[winner]['investments'])
```
